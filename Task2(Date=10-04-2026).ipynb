```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Styling for better visuals (unique touch)
sns.set(style="whitegrid")
```

```python
##Step 1: Load Dataset
```

```python
df = pd.read_csv("student_dataset.csv")

# Display sample data
df.head()
```

```python
##Step 2: Basic Information
```

```python
print("Dataset Info:\n")
df.info()

print("\nStatistical Summary:\n")
df.describe()
```

```python
## Step 3: Distribution Analysis (Histograms)
```

```python
df.hist(figsize=(10,8), bins=20)
plt.suptitle("Feature Distributions", fontsize=14)
plt.show()
```

```python
##Step 5: Feature Relationships (Pairplot)
```

```python
sns.pairplot(df, hue='Gender')
plt.show()
```

```python
## Step 6: Correlation Matrix
```

```python
plt.figure(figsize=(8,5))

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")

plt.title("Correlation Heatmap")
plt.show()
```

```python
## Step 7: Pattern Analysis
```

```python
# Study hours vs Math score
sns.scatterplot(x='Study_Hours', y='Math_Score', data=df)
plt.title("Study Hours vs Math Score")
plt.show()
```

```python
## Step 8: Category Comparison
```

```python
sns.boxplot(x='Gender', y='Math_Score', data=df)
plt.title("Gender vs Math Score")
plt.show()
```

```python
## 🧠 Step 9: Insights & Observations

#Students who study more tend to score higher marks  
#Attendance shows a positive impact on performance  
#Strong correlation between reading and writing scores  
#Gender-based differences are minimal  

#These patterns help in understanding student performance trends
```
